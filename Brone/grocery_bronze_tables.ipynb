###  **Bronze Layer Tables Creation**

**Creation of holidays_events Table**

```python
%py
try:
    spark.sql("USE CATALOG grocery_catalog")
    spark.sql("CREATE SCHEMA IF NOT EXISTS bronze")
    spark.sql("USE SCHEMA bronze")

    spark.sql("DROP TABLE IF EXISTS grocery_catalog.bronze.holidays_events")

    spark.sql("""
        CREATE TABLE grocery_catalog.bronze.holidays_events (
            date INT,
            type STRING,
            locale STRING,
            locale_name STRING,
            description STRING,
            transferred BOOLEAN
        )
        USING PARQUET
        LOCATION 's3://grocery-sales-data-lake-p2/parquet/holidays_events/'
    """)

    print("Bronze table created successfully")

except Exception as e:
    print("Error occurred while creating bronze table")
    raise e
```
